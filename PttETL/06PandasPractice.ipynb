```python
import pandas as pd
import glob
glob.glob('./BabyMother/*.txt')
```

```python
data = list()
for filePath in glob.glob('./BabyMother/*.txt'):
    with open(filePath, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')
        data.append(tmpList[1:-1])
data
```

```python
columns = [c.split(': ')[0] for c in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(s):
    output = s.split(': ')[1]
    return output
```

```python
#df的apply函數:對一個欄位做過濾 (pandas只能一個) 只能自己寫迴圈
df['推'].apply(columnFilter)
```

```python
# 用copy才會有一個新的記憶體位置
df2 = df.copy()
```

```python
df2
```

```python
#對df取值看長相:欄位名稱
for c in df:
    print(c)
```

```python
#將所有欄位套用
for c in df:
    df[c] = df[c].apply(columnFilter)
```

```python
df
```

```python
df2
```

```python
#使用lambda去過濾，簡單邏輯(一句內)用lambda可讀性較高
df2['推'].apply(lambda s: s.split(': ')[1])
```

```python
for c in df2:
    df2[c] = df2[c].apply(lambda s: s.split(': ')[1])
```

```python
df2
```

```python
#儲存成CSV index=0使index那行不出現 header=None使欄位不出現
df.to_csv('./BabyMother.csv', index=0, header=None)
```
